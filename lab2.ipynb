```python
df = df[(df['PS Hydro  ']>0)] # filter data by nonzero PS Hydro values
df.head()
```

### I selected and viewed the Solar column, which was mostly 0 values. I then sorted by PS Hydro, which had two trailing whitespaces that I addressed. Solar didn't have this problem, interestingly.
```python
df["Hydro Total"] = df["Conv. Hydro  "] + df["PS Hydro  "]
sum_hydro_by_year = df.groupby('Year')["Hydro Total"].sum() # creates a subframe with the sum of all hydro values
sum_hydro_by_year.head()
```

```python
sum_hydro_by_year.plot(ylabel="Energy (GWh)", title="Total Hydro Energy Gen. By Year")
```

## This line graph shows the evolution of hydro energy production over time. This combines pumped-storage hydroelectric generation with standard hydroelectric generation for a total amount of energy produced from hydro.

```python
sum_hydro_by_year.plot.bar(ylabel="Energy (GWh)", title="Total Hydro Energy Gen. By Year")
```

## Since I used the same variable (total hydroelectric energy) for this bar chart, the trend looks about the same. This bar chart shows the true proportions between each amount of energy instead of just the trend at the top of the bar. In both, it is clear that there was a dip in hydroelectric generation around 2001, but it recovered soon after and has been overall increasing since then.
